```python
import pandas as pd
import numpy as np
```

```python
chroms = ["chr1", "chr10", "chr10", "chr15", "chr19", "chr2", "chr22", "chr7"]
start = [6000, 13500, 9620, 7460, 4100, 3800, 4225, 9875]
end = [6100, 13550, 9700, 7560, 4200, 3850, 4275, 9975]

locs = pd.DataFrame({"chroms":chroms, "start":start, "end":end})
locs
```

```python
bin_locs = []
for i in locs.values:
    chrom = i[0]
    for j in range(i[1]*10000+100, i[2]*10000, 100):
        bin_locs.append([chrom, j-100, j])
```

```python
bin_locs = pd.DataFrame(bin_locs)
bin_locs
```

```python
bin_locs.to_csv('../data/circosLoci.bed', sep="\t", header=None, index=None)
```
